# Q1. Display Unique Values


```
import pandas as pd

df = pd.DataFrame({"City": ["Bhopal", "Indore", "Bhopal", "Delhi"]})
print(df["City"].unique())
```

# Q2. Count Unique Values


```
import pandas as pd

df = pd.DataFrame({"City": ["Bhopal", "Indore", "Bhopal", "Delhi"]})
print(df["City"].nunique())
```

# Q3. Value Counts


```
import pandas as pd

df = pd.DataFrame({"City": ["Bhopal", "Indore", "Bhopal", "Delhi"]})
print(df["City"].value_counts())
```

# Q4. Group By City


```
import pandas as pd

df = pd.DataFrame({
    "City": ["Bhopal", "Bhopal", "Delhi"],
    "Marks": [80, 90, 70]
})
print(df.groupby("City")["Marks"].mean())
```

# Q5. Sort Values


```
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df.sort_values("Marks"))
```

# Q6. Sort Values Descending


```
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df.sort_values("Marks", ascending=False))
```

# Q7. Filter Records


```
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df[df["Marks"] > 75])
```

# Q8. Multiple Conditions


```
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "Marks": [85, 72, 91, 68]
})
print(df[(df["Marks"] > 70) & (df["Age"] > 20)])
```

# Q9. Group By Department


```
import pandas as pd

df = pd.DataFrame({
    "Employee": ["Aman", "Rahul", "Neha", "Priya"],
    "Department": ["HR", "IT", "IT", "HR"],
    "Salary": [30000, 70000, 50000, 20000]
})
print(df.groupby("Department")["Salary"].sum())
```

# Q10. Complete Practice Question


```
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})

# 1. Show unique cities
print("Unique Cities:", df["City"].unique())

# 2. Count city frequency
print("\nCity Frequency:")
print(df["City"].value_counts())

# 3. Filter marks > 80
print("\nStudents with Marks > 80:")
print(df[df["Marks"] > 80])

# 4. Sort by marks
print("\nSorted by Marks:")
print(df.sort_values("Marks"))
```
