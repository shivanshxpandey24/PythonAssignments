## Q1. Sort Age in Ascending Order


```python
import pandas as pd

df = pd.DataFrame({"Age": [25, 18, 30, 22]})
print(df["Age"].sort_values().values)
```

## Q2. Sort Marks in Descending Order


```python
import pandas as pd

df = pd.DataFrame({"Marks": [70, 95, 80, 60]})
print(df["Marks"].sort_values(ascending=False).values)
```

## Q3. Find Mean Age


```python
import pandas as pd

df = pd.DataFrame({"Age": [20, 25, 30, 25]})
print(df["Age"].mean())
```

## Q4. Find Median Marks


```python
import pandas as pd

df = pd.DataFrame({"Marks": [50, 60, 70, 80, 90]})
print(df["Marks"].median())
```

## Q5. Find Mode


```python
import pandas as pd

df = pd.DataFrame({"Age": [20, 25, 25, 30, 25]})
print(df["Age"].mode()[0])
```

## Q6. Find Maximum Score


```python
import pandas as pd

df = pd.DataFrame({"Score": [78, 90, 67, 88]})
print(df["Score"].max())
```

## Q7. Find Minimum Score


```python
import pandas as pd

df = pd.DataFrame({"Score": [78, 90, 67, 88]})
print(df["Score"].min())
```

## Q8. Count City Frequency


```python
import pandas as pd

df = pd.DataFrame({"City": ["Bhopal", "Indore", "Bhopal", "Jabalpur"]})
print(df["City"].value_counts())
```

## Q9. Find Percentage Frequency


```python
import pandas as pd

df = pd.DataFrame({"City": ["A", "A", "B", "C"]})
print(df["City"].value_counts(normalize=True) * 100)
```

## Q10. Sort DataFrame by Salary


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj", "Riya"],
    "Salary": [30000, 50000, 40000]
})
df_sorted = df.sort_values("Salary", ascending=False)
print(df_sorted["Name"].values)
```

## Q11. Find Correlation


```python
import pandas as pd

df = pd.DataFrame({
    "Math": [10, 20, 30, 40],
    "Science": [15, 25, 35, 45]
})
print(df["Math"].corr(df["Science"]))
```

## Q12. Find Covariance


```python
import pandas as pd

df = pd.DataFrame({
    "Math": [10, 20, 30, 40],
    "Science": [15, 25, 35, 45]
})
print(round(df["Math"].cov(df["Science"]), 2))
```

## Q13. Sort by Two Columns


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "City": ["Indore", "Bhopal", "Bhopal", "Indore"],
    "Age": [21, 22, 20, 23]
})
print(df.sort_values(["City", "Age"]))
```

## Q14. Student Statistics


```python
import pandas as pd

df = pd.DataFrame({"Marks": [60, 70, 80, 90]})
print("Mean =", df["Marks"].mean())
print("Median =", df["Marks"].median())
print("Max =", df["Marks"].max())
print("Min =", df["Marks"].min())
```

## Q15. Mini Practice


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "Score": [85, 72, 91, 68, 95],
    "Height": [5.7, 5.9, 5.5, 5.8, 6.0]
})

# 1. Sort by Score (Descending)
print("--- Sorted by Score (Descending) ---")
print(df.sort_values("Score", ascending=False))

# 2. Find Mean Age
print("\nMean Age:", df["Age"].mean())

# 3. Find Median Height
print("Median Height:", df["Height"].median())

# 4. Show Score Frequency
print("\nScore Frequency:")
print(df["Score"].value_counts())
```
