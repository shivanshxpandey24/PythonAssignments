
# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
import pandas as pd

data = {
    "Name": ["Aman", None, "Neha"],
    "Marks": [80, 75, None]
}
df = pd.DataFrame(data)

# Filling missing names with 'Unknown' and missing marks with 0
df["Name"] = df["Name"].fillna("Unknown")
df["Marks"] = df["Marks"].fillna(0)

print("Missing values replaced:")
print(df)
````


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
import pandas as pd

data = {
    "City": ["BPL", "IND", "BPL"]
}
df = pd.DataFrame(data)

df["City"] = df["City"].replace({"BPL": "Bhopal", "IND": "Indore"})
print(df)
````


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
import pandas as pd

data = {
    "Name": ["Aman", "Rahul", "Aman", "Neha"]
}
df = pd.DataFrame(data)

print("Duplicate rows check:")
print(df.duplicated())

print("\nAfter removing duplicates:")
df_cleaned = df.drop_duplicates()
print(df_cleaned)
````
