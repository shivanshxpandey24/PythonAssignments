# Q1. Array Attributes

## Problem Statement
Create the following NumPy array:

```python
[10, 20, 30, 40, 50]
```

Print:
- Shape
- Number of Dimensions
- Size
- Data Type

## Sample Output
```python
Shape: (5,)
Dimensions: 1
Size: 5
dtype: int64
```


````python
import numpy as np

arr=[10,20,30,40,50]
arr=np.array(arr)

#.shape --> ratio of matrix
print(arr.shape)

#.ndim --> dimention if array
print(arr.ndim)

#.size --> no. of elements
print(arr.size)

#.dtype --> data type inside your array
print(arr.dtype)
````
