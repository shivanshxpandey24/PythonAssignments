# Q1. Create a Student DataFrame

# Q2. Display Selected Columns


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
print(df[["Name", "Marks"]])
```

# Q3. Filter Students


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
print(df[df["Marks"] > 75])
```

# Q4. Sort by Marks


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
print(df.sort_values("Marks", ascending=False))
```

# Q5. Average Marks by City


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
print(df.groupby("City")["Marks"].mean())
```

# Q6. Count Students by City


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
print(df["City"].value_counts())
```

# Q7. Create Result Column


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Age": [21, 22, 20, 23, 21],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print(df)
```

# Q8. Remove Duplicate Rows


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Aman", "Neha"],
    "Age": [21, 22, 21, 20],
    "City": ["Bhopal", "Indore", "Bhopal", "Delhi"],
    "Marks": [85, 72, 85, 91]
})
df = df.drop_duplicates()
print(df)
```

# Q9. Fill Missing Values


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, np.nan, 91, np.nan]
})
df["Marks"] = df["Marks"].fillna(0)
print(df)
```

# Q10. Final Practice


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
})

# 1. Display students with Marks > 80
print("--- Students with Marks > 80 ---")
print(df[df["Marks"] > 80])

# 2. Sort the DataFrame by Marks
print("\n--- Sorted by Marks ---")
print(df.sort_values("Marks"))

# 3. Find the average Marks of each City
print("\n--- Average Marks by City ---")
print(df.groupby("City")["Marks"].mean())
```
