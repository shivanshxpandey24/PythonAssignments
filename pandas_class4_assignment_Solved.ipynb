# Pandas Class 4 Assignment


# Q1. Select Single Column


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha"],
    "Age": [21, 22, 20]
})
print(df["Name"])
```

# Q2. Select Multiple Columns


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha"],
    "Age": [21, 22, 20],
    "Marks": [85, 72, 91]
})
print(df[["Name", "Marks"]])
```

# Q3. Filter Students


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df[df["Marks"] > 80])
```

# Q4. Filter by City


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal"],
    "Marks": [85, 72, 91, 68]
})
print(df[df["City"] == "Bhopal"])
```

# Q5. Find Maximum Marks


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df["Marks"].max())
```

# Q6. Find Minimum Age


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23]
})
print(df["Age"].min())
```

# Q7. Sort Data


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, 72, 91, 68]
})
print(df.sort_values("Marks"))
```

# Q8. Value Counts


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal"]
})
print(df["City"].value_counts())
```

# Q9. Group By City


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal"],
    "Marks": [85, 72, 91, 68]
})
print(df.groupby("City")["Marks"].mean())
```

# Q10. Multiple Conditions


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "Marks": [85, 72, 91, 68]
})
print(df[(df["Marks"] > 70) & (df["Age"] > 20)])
```
