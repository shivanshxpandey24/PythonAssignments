## Q1. Find Missing Values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Riya", None],
    "Age": [21, np.nan, 25]
})
print(df.isna())
```

## Q2. Count Total Missing Values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", None, "Raj"],
    "Age": [20, None, 25]
})
print(df.isnull().sum())
```

## Q3. Fill Missing Age with 0

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj"],
    "Age": [20, None]
})
df["Age"] = df["Age"].fillna(0)
print(df)
```

## Q4. Fill Missing City with 'Unknown'

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj"],
    "City": ["Bhopal", None]
})
df["City"] = df["City"].fillna("Unknown")
print(df)
```

## Q5. Remove Rows Having Null Values

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj", None],
    "Age": [20, 25, 30]
})
df = df.dropna()
print(df)
```

## Q6. Convert Age Column into Float

```python
import pandas as pd

df = pd.DataFrame({"Age": [20, 25, 30]})
df["Age"] = df["Age"].astype(float)
print(df["Age"].tolist())
```

## Q7. Replace Delhi with New Delhi

```python
import pandas as pd

df = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Delhi"]
})
df["City"] = df["City"].replace("Delhi", "New Delhi")
print(df["City"].tolist())
```

## Q8. Replace Score 88 with 100

```python
import pandas as pd

df = pd.DataFrame({
    "Score": [88, 90, 88, 75]
})
df["Score"] = df["Score"].replace(88, 100)
print(df["Score"].tolist())
```

## Q9. Drop City Column

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj"],
    "City": ["Bhopal", "Indore"]
})
df = df.drop(columns=["City"])
print(df)
```

## Q10. Drop Row at Index 1

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj", "Riya"]
})
df = df.drop(index=1)
print(df)
```

## Q11. Check Duplicate Rows

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj", "Aman"]
})
print(df.duplicated().tolist())
```

## Q12. Remove Duplicate Rows

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman", "Raj", "Aman"]
})
df = df.drop_duplicates()
print(df)
```

## Q13. Mini Data Cleaning Task

```python
import pandas as pd

data = {
    "Name": ["Aman", "Raj", "Aman", None],
    "Age": [20, None, 20, 25],
    "City": ["Bhopal", "Delhi", "Bhopal", "Indore"]
}
df = pd.DataFrame(data)

# Step 1: Fill missing Age with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Step 2: Remove duplicate rows
df = df.drop_duplicates()

# Step 3: Display final DataFrame
print(df)
```
